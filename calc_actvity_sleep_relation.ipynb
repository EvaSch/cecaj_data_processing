### Sleep score
based on Dr. Oz
http://blog.doctoroz.com/oz-experts/calculating-your-perfect-bedtime-and-sleep-efficiency

NOTE: number of awakenings are not included in the score


The labels for the sleep quality are also based on Dr. Oz classification.
* <= 75% bad sleep -> own classification -> 3
*  \> 75% sufficient sleep -> own classification -> 2
*  \> 85% normal sleep -> 1
*  \> 90% good sleep -> 0

```python
######### config section #########
activity_data_file = "OneYearFitBitDataEdited.csv"
sleep_data_file = "OneYearFitBitDataSleep.csv"
sleep_data_score_output_file = 'OneYearFitBitDataSleepScore.csv'
```

```python
import pandas as pd
from IPython.display import display, HTML
```

```python
def main():
    sleep_data = pd.read_csv(sleep_data_file, sep = ',')
    sleep_score_df = calc_sleep_score(sleep_data)
    sleep_score_df = create_sleep_score_label(sleep_score_df)
    display(sleep_score_df)
    sleep_score_df.to_csv(sleep_data_score_output_file, index=False)
    
    return 
```

```python
def calc_sleep_score(sleep_data):
    
    sleep_data_col = (sleep_data['MinutesOfSleep'] - sleep_data['MinutesOfBeingAwake'])/ \
                      sleep_data['LengthOfRestInMinutes']
    
    sleep_score_df = pd.concat([sleep_data, sleep_data_col], axis=1)
    sleep_score_df = sleep_score_df.rename(columns = {0: 'SleepScore'}) # adds column header
    
    return sleep_score_df
```

```python
def create_sleep_score_label(sleep_score_df):
    sleep_scores = sleep_score_df['SleepScore']
    sleep_labels = pd.DataFrame(columns=['sleep_labels'])

    for score in sleep_scores:
        if score > 0.9:
            sleep_labels = sleep_labels.append(pd.Series(['0'],index = ['sleep_labels']), ignore_index=True)
        elif score > 0.85:
            sleep_labels = sleep_labels.append(pd.Series(['1'],index = ['sleep_labels']), ignore_index=True)
        elif score > 0.75:
            sleep_labels = sleep_labels.append(pd.Series(['2'],index = ['sleep_labels']), ignore_index=True)
        elif score <= 0.75:
            sleep_labels = sleep_labels.append(pd.Series(['3'],index = ['sleep_labels']), ignore_index=True)
        else:
            sleep_labels = sleep_labels.append(pd.Series(['NaN'],index = ['sleep_labels']), ignore_index=True)
            
            
    sleep_score_df = pd.concat([sleep_score_df, sleep_labels], axis=1)   
    
    
    return sleep_score_df
```

```python
if __name__ == '__main__':
    main()
```
